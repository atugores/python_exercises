# Multiplots

Use multiplots and logarithmical scale

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

fig = plt.figure(1)

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')


ax1 = plt.subplot(221)  # if you want to change the axes, get them
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
ax1.set_yticks([-1,0,1])

plt.subplot(222)
plt.plot(t1, f(t1), 'go', t2, f(t2), 'k')

# LOG SCALE
plt.yscale('log')


# adjust and the location of axes needs to be adjusted
fig.subplots_adjust(hspace=0.2)
plt.tight_layout()
```
