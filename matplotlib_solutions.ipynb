```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

## Generate some data

Create an array, x, with 256 equidistant values in [-pi, pi].

```python
x = np.linspace(-np.pi, np.pi, 100, endpoint=True)
```

Create an array, c, with the values: cos(x)

Create an array, s, with the values: sin(x)

```python
c, s = np.cos(x), np.sin(x)
```

## Plotting

Plot sin(x) and cos(x) in the same plot.

```python
plt.plot(x,c)
plt.plot(x,s)
```

### Move spines

Set spines to x=0 and y=0

```python
fig, ax = plt.subplots()
ax.set_xlabel('this is x', labelpad=100) # BE CAREFUL WITH LABELS
ax.set_ylabel('this is y', labelpad=150) # BE CAREFUL WITH LABELS
ax.set_title('some title here', y=1.08)
plt.plot(x,c, linewidth=0.8, linestyle='-')
plt.plot(x,s, linewidth=0.2, linestyle='-')
ax.grid(color='lightgray', alpha=0.7)

plt.ylim(-1.2, 1.2)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

# decide which spines to show
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#ax.spines['left'].set_color('none')
#ax.spines['bottom'].set_color('blue')

# set position to the visible ones
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))


ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')
```

Let’s add a legend. Try using different locations and different styles (shadow, box, ...)

```python
fig, ax = plt.subplots()
ax.set_xlabel('this is x', labelpad=100)
ax.set_ylabel('this is y', labelpad=150)
ax.set_title('some title here', y=1.08)
plt.plot(x,c, linewidth=0.8, linestyle='-', label="cosine")
plt.plot(x,s, linewidth=0.2, linestyle='-', label="sine")
ax.grid(color='lightgray', alpha=0.7)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

# plt.legend(loc='upper left') # upper lower center - right left center
plt.legend(loc='upper left', frameon=False) # , framealpha=0.7
```

Annotate some interesting points using the annotate command. We chose the $2\pi/3$ value and we want to annotate both the sine and the cosine. First draw a marker on the curve as well as a straight dotted line. Then, use the annotate command to display some text with an arrow.

```python
fig, ax = plt.subplots()
ax.set_xlabel('this is x', labelpad=100)
ax.set_ylabel('this is y', labelpad=150)
ax.set_title('some title here', y=1.08)
plt.plot(x,c, linewidth=0.8, linestyle='-', label="cosine")
plt.plot(x,s, linewidth=1, linestyle='-', label="sine")
ax.grid(color='lightgray', alpha=0.7)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

plt.legend(loc='upper left', framealpha=0.7)

ax.set_ylim(-2, 2)
ax.set_xlim(-2, 4)


t = 2 * np.pi / 3
plt.plot([t, t],[0, np.sin(t)], color='green', linewidth=1.5, linestyle="--")
plt.scatter([t, ],[np.sin(t), ], 30, color='green', alpha=0.2)
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=1.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 20, color='blue')
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
```

The tick labels are now hardly visible because of the blue and red lines. We can make them bigger and we can also adjust their properties such that they’ll be rendered on a semi-transparent white background. This will allow us to see both the data and the labels.

```python
fig, ax = plt.subplots()
ax.set_xlabel('this is x', labelpad=100)
ax.set_ylabel('this is y', labelpad=90)
ax.set_title('some title here', y=1.08)
plt.plot(x,c, linewidth=0.8, linestyle='-', label="cosine")
plt.plot(x,s, linewidth=0.2, linestyle='-', label="sine")
ax.grid(color='lightgray', alpha=0.7)

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
#ax.spines['left'].set_color('none')
#ax.spines['bottom'].set_color('blue')


ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.xaxis.set_ticks_position('bottom')

plt.legend(loc='upper left', framealpha=0.4)

ax.set_ylim(-2, 2)
ax.set_xlim(-2, 4)


t = 2 * np.pi / 3
plt.plot([t, t],[0, np.sin(t)], color='green', linewidth=1.5, linestyle="--")
plt.scatter([t, ],[np.sin(t), ], 30, color='green', alpha=0.2)
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
             xy=(t, np.sin(t)), xycoords='data',
             xytext=(+10, +30), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=1.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 20, color='blue')
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
             xy=(t, np.cos(t)), xycoords='data',
             xytext=(-90, -50), textcoords='offset points', fontsize=16,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
```
